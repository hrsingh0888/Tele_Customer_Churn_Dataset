# **Churn Visualization**

```python
plt.figure(figsize = (6,5))
plt.bar(list(df['Churn'].value_counts().keys()),df['Churn'].value_counts())
plt.title('Churn')
plt.show()
```

```python
plt.figure(figsize = (5,5))
plt.bar(list(df['gender'].value_counts().keys()), df['Churn'].value_counts())
plt.title('Churn Vs Gender')
plt.xlabel('Gender')
plt.ylabel('Churn')
plt.show()
```

```python
plt.figure(figsize = (5,5))
plt.bar(list(df['Partner'].value_counts().keys()), df['Churn'].value_counts())
plt.title('Churn Vs Partner')
plt.xlabel('Partner')
plt.ylabel('Churn')
plt.show()
```

```python
plt.figure(figsize = (5,5))
churn_by_payment_method = df.groupby('PaymentMethod')['Churn'].value_counts()
categories = churn_by_payment_method.index.get_level_values('PaymentMethod').unique()
churn_counts = churn_by_payment_method.groupby('PaymentMethod').sum()

plt.bar(categories, churn_counts)
plt.title('Churn Vs PaymentMethod')
plt.xlabel('PaymentMethod')
plt.ylabel('Churn Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (5,5))
plt.bar(list(df['Dependents'].value_counts().keys()), df['Churn'].value_counts())
plt.title('Churn Vs Dependents')
plt.xlabel('Dependents')
plt.ylabel('Churn')
plt.show()
```

```python
plt.figure(figsize = (5,5))
# Group churn counts by contract type
churn_by_contract = df.groupby('Contract')['Churn'].value_counts()

# Extract contract types and corresponding churn counts
contracts = churn_by_contract.index.get_level_values('Contract').unique()
churn_counts = churn_by_contract.groupby('Contract').sum()

plt.bar(contracts, churn_counts)
plt.title('Churn Vs Contract')
plt.xlabel('Contract')
plt.ylabel('Churn Count')
plt.show()
```

```python
plt.figure(figsize = (5,5))
plt.plot(df['tenure'])
plt.title('Tenure')
plt.xlabel('index')
plt.ylabel('tenure')
plt.show()
```
