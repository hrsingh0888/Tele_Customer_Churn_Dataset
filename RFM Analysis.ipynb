```python
import pandas as pd
from datetime import datetime
```

```python
data_type = {
    'customerID': str
}
```

```python
df = pd.read_csv('/content/Telco_Customer_Churn_Dataset(s).csv', dtype = data_type, parse_dates = ['tenure'])
```

```python
columns = ['tenure', 'customerID', 'Contract', 'MonthlyCharges']
df_dataset = df[columns]
```

```python
today_date = pd.to_datetime('2022-01-01')
```

```python
Rfm_datasets = df_dataset.groupby('customerID').agg({
    'tenure': 'max',
    'Contract': 'first',
    'MonthlyCharges': 'mean'
    })
```

```python
Rfm_datasets
```

```python
Rfm_datasets.rename(columns = {
    'tenure': 'recency',
    'Contract': 'frequency', 'MonthlyCharges':'monetary'
    },
    inplace = True
)
```

```python
Rfm_datasets
```

```python
r = pd.qcut(Rfm_datasets['recency'], q = 5, labels = range(5, 0, -1))
f = pd.qcut(Rfm_datasets['frequency'], q = 5, labels = range(1, 6))
m = pd.qcut(Rfm_datasets['monetary'], q = 5, labels = range(1, 6))
```

```python
Rfm_datasets['recency'] = pd.to_numeric(Rfm_datasets['recency'])
```
